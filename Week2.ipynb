```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
img = cv.imread('/content/drive/MyDrive/PCVK/kitten01.jpg')
plt.imshow(img)

#Convert to RGB
img2 = img
img3 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img3)

#CMAP
img_gray = cv.imread('/content/drive/MyDrive/PCVK/kitten01.jpg', cv.IMREAD_GRAYSCALE)
plt.imshow(img_gray, cmap='magma')

#Resize
img4 = cv.resize(cv.cvtColor(img, cv.COLOR_BGR2RGB), (512, 1024))
plt.imshow(img4)

#Flip
img5 = cv.flip(cv.cvtColor(img, cv.COLOR_BGR2RGB), 0)
cv.imwrite('/content/drive/MyDrive/PCVK/kitten01_flip.jpg', img5)
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
ax.imshow(img5)

#Create black image
black_img = np.zeros(shape=[512, 512, 3], dtype=np.int16)
plt.imshow(black_img)

#Create rectangle
cv.rectangle(black_img, pt1=(384, 0), pt2=(510, 150), color=(0, 255, 0), thickness=10)
plt.imshow(black_img)

#Create rectangle 2
cv.rectangle(black_img, pt1=(200, 200), pt2=(300, 300), color=(0, 0, 255), thickness=15)
plt.imshow(black_img)

#Create circle
cv.circle(black_img, center=(100,100), radius=50, color=(255, 0, 0), thickness=8)
plt.imshow(black_img)

#Create line
cv.line(black_img, pt1=(0, 0), pt2=(512, 512), color=(255, 255, 255), thickness=5)
plt.imshow(black_img)

#Add text and font
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img, text='Hello', org=(10, 500), fontFace=font, fontScale=4, color=(255, 255, 0), thickness=2, lineType=cv.LINE_AA)
plt.imshow(black_img)


#Black image with int data
black_img2 = np.zeros(shape=[512, 512, 3], dtype=np.int32)
plt.imshow(black_img2)

#inisialisasi NumPy array int32
vertices = np.array([[100, 300], [200, 200], [400, 300], [200, 400]], dtype=np.int32)
vertices

pts = vertices.reshape((-1, 1, 2))
pts

# #Create polygon
cv.polylines(black_img2, [pts], isClosed=True, color=(255, 0, 0), thickness=5)
plt.imshow(black_img2)
```

```python
#1. Figsize
img2 = img
img3 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.figure(figsize=(5,5))
plt.imshow(img3)
plt.show()

#2. Channel Red-Blue and Green-Blue
red_blue = img.copy()
red_blue[:, :, 1] = 0  # Set saluran hijau menjadi 0 (hanya tinggalkan saluran merah dan biru)

# Channel Green-Blue
green_blue = img.copy()
green_blue[:, :, 2] = 0  # Set saluran merah menjadi 0 (hanya tinggalkan saluran hijau dan biru)

red_blue_rgb = cv.cvtColor(red_blue, cv.COLOR_BGR2RGB)
green_blue_rgb = cv.cvtColor(green_blue, cv.COLOR_BGR2RGB)

# Menggabungkan gambar secara horizontal
final_frame = np.hstack((red_blue_rgb, green_blue_rgb))

# Menampilkan gambar yang digabungkan
plt.imshow(final_frame)
plt.axis('off')
plt.show()

#3. Crop Image
# Memotong bagian tertentu dari citra
cropped_image = img[20:115, 25:120]

# Menampilkan citra hasil pemotongan
plt.imshow(cv.cvtColor(cropped_image, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

#4. Crop Image only Red channel
# Melakukan cropping gambar: baris 5-30 dan semua kolom
cropped_img_2 = img3[5:30, :, 0]  # Hanya channel Red (indeks 0)

# Menampilkan gambar hasil cropping
plt.imshow(cropped_img_2)
plt.axis('off')
plt.show()

#5. Random Rectangle
import random

# # Buat gambar kosong (hitam)
image = np.zeros(shape=(400,400,3),dtype=np.int32)

# # Buat 5 kotak dengan ukuran dan warna acak
for _ in range(5):
    # Tentukan ukuran acak antara 20 dan 100
    width = random.randint(20, 100)
    height = random.randint(20, 100)

    # Tentukan posisi acak
    x = random.randint(0,400 - width)  # Maksimal x adalah 400 agar kotak tidak melewati batas gambar
    y = random.randint(0,400 - height)  # Maksimal y adalah 400 agar kotak tidak melewati batas gambar

    # Tentukan warna acak
    color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))

    # Gambar kotak dengan ukuran, posisi, dan warna yang sudah ditentukan
    cv.rectangle(image, (x, y), (x + width, y + height), color, 2)

cv2_imshow(image)

#6. Flip image
flip_image = cv.flip(cv.cvtColor(img,cv.COLOR_BGR2RGB), 0)

plt.imshow(flip_image)

#7. Activity Image
activity_image = cv.imread('/content/drive/MyDrive/PCVK/SportifYan.JPG')
myPhoto = cv.cvtColor(activity_image,cv.COLOR_BGR2RGB)

cv.rectangle(myPhoto,pt1=(750,1500),pt2=(1350,2100), color=(0,255,0), thickness=20)
cv.circle(myPhoto,center=(1050,1800),radius=300, color=(255,0,0), thickness=20)
plt.imshow(myPhoto)

#8. Rectangle kitten
# Membaca gambar
kitten = img.copy()

# Menentukan ukuran dan warna kotak (rectangle)
rect_width, rect_height = 200, 200
blue_color = (255, 0, 0)

# Menambahkan kotak (rectangle) pada sudut bawah kiri di saluran biru (Blue channel)
# Pada array, slicing dilakukan untuk memilih area tertentu dari gambar.
kitten[-rect_height:, :rect_width, 0] = blue_color[0]  # Menetapkan hanya nilai biru pada saluran 0

# Menampilkan gambar yang telah dimodifikasi
cv2_imshow(kitten)

#9. Complete kitten
# Membaca gambar
kitten = img.copy()

# Menentukan ukuran dan warna kotak (rectangle)
rect_width, rect_height = 200, 200
blue_color = (255, 0, 0)

# Menambahkan kotak (rectangle) pada sudut bawah kiri di saluran biru (Blue channel)
# Pada array, slicing dilakukan untuk memilih area tertentu dari gambar.
kitten[-rect_height:, :rect_width, 0] = blue_color[0]  # Menetapkan hanya nilai biru pada saluran 0

font = cv.FONT_HERSHEY_COMPLEX
cv.putText(kitten,text='Kitten01',org=(200,500),fontFace=font,fontScale=2,color=(0,0,0),thickness=2,lineType=cv.LINE_AA)
plt.imshow(kitten)
```
